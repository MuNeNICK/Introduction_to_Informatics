```python
import networkx as nx
G=nx.Graph()
G.add_nodes_from([1,2,3,4,5])
G.add_edges_from([(1,2),(1,5),(2,3),(2,5),(3,4),(4,5)])
nx.draw_networkx(G)
```

```python
A=nx.adjacency_matrix(G)
M=nx.incidence_matrix(G)
print('A=',A.toarray())
print('M=',M.toarray())
```
