```python
import networkx as nx
GR=nx.grid_2d_graph(3,3)
GR.add_edges_from([((0,1),(1,2))])
nx.draw_networkx(GR,pos={v:v for v in GR.nodes()})
nx.is_eulerian(GR)
```

```python
GR.add_edges_from([((1,0),(2,1))])
nx.draw_networkx(GR,pos={v:v for v in GR.nodes()})
nx.is_eulerian(GR)
```

```python
ee=nx.eulerian_circuit(GR)
for (i,j) in ee:
  print(i, end='->')
```

```python
GR=nx.complete_graph(5)
nx.draw_networkx(GR)
nx.is_eulerian(GR)
```
