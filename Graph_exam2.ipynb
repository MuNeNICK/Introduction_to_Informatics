```python
import networkx as nx
G=nx.complete_graph(5)
nx.draw_networkx(G)
```

```python
m,n=3,4
G=nx.complete_bipartite_graph(m,n)
nx.draw_networkx(G)
```

```python
m,n=3,4
G=nx.complete_bipartite_graph(m,n)
p={}
for i in range(m):
  p[i]=(0,i)
for i in range(n):
  p[m+i]=(1,i)
nx.draw_networkx(G,pos=p)
```
